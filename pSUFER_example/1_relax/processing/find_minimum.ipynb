```python
import pandas as pd
import os
from glob import glob

def main():
    
    pdblist=[]
    pdbfiles=glob("/projects/jupiter/abfeich/new_proteins/LSO/1_relax/output/*/*_scores.sc")
    
    for pdb in pdbfiles:    
        print(pdb)
        score_file_all = open(pdb).readlines()
    
        headers =  score_file_all[0].split()[1:]
        score_data = [line.split()[1:] for line in score_file_all[1:]] #score data
        for i in range(len(score_data)):
            print(len(score_data[i]))

        print("there are", len(headers), " headers")
#        print(score_data)
        
        score_df = pd.DataFrame(score_data, columns = headers, dtype=float)
        score_df = score_df.sort_values(by=["total_score"])
#        print(score_df.columns)

        minimum = score_df.iloc[0]

#        print(minimum["description"])
        pdblist.append(minimum["description"])
    
    with open("/projects/jupiter/abfeich/new_proteins/LSO/1_relax/minscores_all", 'w') as f:
        for p in pdblist:
            f.write(f'{p}\n')
    f.close()

main()
```
